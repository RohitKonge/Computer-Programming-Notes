```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("Classified Data", index_col=0)
df1.info()
df1 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df1.drop(('TARGET CLASS') , axis = 1 ))
scaled_features = scaler.transform(df1.drop(('TARGET CLASS') , axis = 1 ))
scaled_features
```

```python
df1_feat = pd.DataFrame(scaled_features, columns=df1.columns[:-1])
df1_feat
```

```python
from sklearn.model_selection import train_test_split
X = df1_feat
y = df1['TARGET CLASS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors= 1)              # No of neighbors we want
KNN.fit(X_train, y_train)
```

```python
prediction =  KNN.predict(X_test)
prediction
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(prediction, y_test))

print(classification_report(prediction, y_test))
```

```python
# Now we will improve our model by using Elbow Method

error_rate = []             #   Check the model's prediction for a bunch of 'k' values and then get their errors and then 
                            #   get the lowest error among them
for i in range(1,41):
    knn  = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_rate.append(np.mean(pred != y_test))

error_rate
```
