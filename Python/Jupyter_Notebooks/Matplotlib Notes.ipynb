# Functional Plots

```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

x = np.linspace(0,10,30)
plt.xlabel("XLABEL")
plt.ylabel("YLABEL")
plt.title("TITLE")
plt.plot(x,x**2,'-r', label = "x^2")
plt.legend(loc = (0.1,0.2))                # Gave custom Position to Legend
```

# Functional Subplots

```python
plt.subplot(1,2,1)
plt.plot(x,x**3)
plt.subplot(1,2,2)
plt.plot(x**3,x)
```

# Object Oriented Plot

```python
fig1 = plt.figure(1, (11,6), 100, 'y', 'r')
axes1 = fig1.add_axes([0.00,0.00,0.3,0.3])
axes1.set_xlabel("XLABEL")
axes1.set_ylabel("YLABEL")
axes1.set_title("Title")
axes1.plot(x,x**2,label="x,x2")
axes1.plot(x,x**3,label="x,x2")
axes1.legend(loc = 0)
# loc = 0    Decides itself
# loc = 1    Upper Right
# loc = 2    Upper Left
# loc = 3    Lower Left
# loc = 4    Lower Right
# axes1.legend(loc=(0.1,0.1))            For Custom Place

axes2 = fig1.add_axes([0.2,0.3,0.2,0.2])
axes2.plot()
```

# Object Oriented Subplot

```python
fig1, axes1 = plt.subplots(1,3)
plt.tight_layout()
axes1[0].plot(x,x**2)
axes1[1].plot(x,x**3)
axes1[2].plot(x,x)
```



# Appereance Options

```python
fig1 = plt.figure( facecolor = "#c8d3e9")

axes1 = fig1.add_axes([0,0,1,1])
axes1.plot(x,x**2, lw = 1, ls = '-.', alpha = 0.6, marker = 'o')
axes1.set(xlim=(0,6), ylim=(-10,40))
axes1.grid(visible = True)

axes2 = fig1.add_axes([1.5,0,0.5,0.5])
```

# Working with Pandas Dataframes

```python
# Here we plot the data from dataframes by taking the 'numpy array' of the columns as X and Y

df = pd.read_csv("E:/Computer Programming/VS Code Projects/GithubCPNotes/Computer-Programming-Notes/Python/Utilities/EcommercePurchases.csv")
np_arr = df.values

y = np_arr[:20,-1]
x = range(len(y))

fig1 = plt.figure()
axes1 = fig1.add_axes([0,0,1,1])
axes1.text(7.5,20,"The Text")
axes1.text(5,17,r"$\sigma$")                               # Markup texts
axes1.plot(x,y,label="asdf")
axes1.legend()
```

# Bar Charts
