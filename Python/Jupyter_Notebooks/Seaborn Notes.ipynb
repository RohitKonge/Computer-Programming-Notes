```python
import seaborn as sns, numpy as np
%matplotlib inline


#Univariate Graph --> Plotting a distribution of data with 1 variable 
#Bivariate Graph --> Plotting a distribution of data with 2 variables
#KDE            --> Kernal Density Distribution

data1 = sns.load_dataset('tips')

data2 = sns.load_dataset('penguins')

data3 = sns.load_dataset('flights')

data4 = sns.load_dataset('iris')

# data1

# data2

# data3[['month', 'year']][3:10]

data4['species'].unique()
```

# Distribution Plots

# 

# Displot & Rug Plot

# 

```python
sns.displot(data = data1, x = data1['tip'], y = data1['total_bill'], kind = 'hist', height = 5)
sns.rugplot(x = data1['tip'], expand_margins = True, height = 0.05, hue = data1['sex'])
```

# Joint Plot

# 

```python
sns.jointplot(data = data1, x = data1['tip'], y = data1['total_bill'], xlim = (0,8), height = 5, kind = 'hex', color = 'r'
             , space = 0, ratio  = 3)
```

# Pair Plot 

# 

```python
sns.pairplot(data = data1[['tip', 'total_bill']], kind = 'hist', corner  = False, height = 3, aspect = 0.9, diag_kind = 'kde')

#hue is used for differentiating between categorical column i.e haveing only distinct values eg. male/female
```

# Heat Map

```python
# For Heat Map Data Should be in a Matrix Form

# Matrix form is such that the index name and the column name match up so that the cell value shows something that is 
# relevant to both of these

# We make the Matrix form by Pivot Table / Correlation Data

# Using Correlation Data Method

t = data2.corr()
sns.heatmap(t, annot = True, cmap = 'coolwarm', linewidths = 2, square = True)
```

```python
# Using Pivot Table

t = data3.pivot_table('passengers', 'month', 'year')
sns.heatmap(t, cmap = 'coolwarm', linewidths = 2, square = True)
```

# Cluster Map

```python
sns.clustermap(t)   # Gives info on the relation between 2 columns
```

# Pair Grid 

```python
# Pair Plot is a simplified version of Pair Grid

# Pair Plot does a lot of stuff automatically and Pair Grid give you a lot of control

import matplotlib.pyplot as plt

t = sns.PairGrid(data4, height = 2)
# t.map(plt.scatter)                   # this puts scatter plot on every plot of the grid
t.map_diag(sns.histplot, color = 'y')
t.map_upper(sns.kdeplot, color = 'r')
t.map_lower(plt.scatter, color = 'g')
```
